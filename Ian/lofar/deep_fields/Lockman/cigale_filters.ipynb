## Code to break the single filter file into individual files for each filter for CIGALE

```python
import glob
import os
import shutil
```

```python
f = open('data/photz/filters/Lockman-SWIRE_filters.res','r')
filters = []
filters_num = []
for line in f:
    line_split = line.split()
    print(line_split)
    if 'res' in line:
        
        filters_num.append(int(line_split[0]))
        filters.append(line_split[1])
        filt = open('data/photz/filters/{}'.format(line_split[1].replace('res','dat')),'w')
        filt.write('# {}\n'.format(line_split[1].replace('.res','')))
        filt.write('# energy\n')
        continue
    filt.write('{} {}\n'.format(line_split[1],line_split[2]))
```
